```python
#importing neccesary libraries
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#dealing with null values
df.fillna(value = '0')
```

```python
#converting categorical coloumns to numericals
le = LabelEncoder()
data_coloumns = ['Formatted Date' , 'Summary' , 'Precip Type' , 'Daily Summary']
for columns in data_coloumns:
    df[columns] = le.fit_transform(df[columns])
df.head() 
```

```python
#converting the dataframe into array for calculation
data_array = df.to_numpy()

#function for covariance matrix
def cova(data_array):
    
    #calculating mean
    mean = np.mean(data_array , axis = 0)
    
    #calculating the center point
    center = data_array - mean
    
    #calculating the covariance by multiplying it 
    cov_mat = np.dot(center.T , center) / (len(data_array) - 1)
    
    return cov_mat

cov_m = cova(data_array)
print(cov_m)
```

```python
#function for calculating pca
def pca_f(data_array , num_components):
    
    #calling the cov_matrix function
    cov_m = cova(data_array)
    
    #getting the eigen values and eigen vectors
    eigen_values , eigen_vectors = np.linalg.eig(cov_m)
    
    #selecting the top eigen values
    sor_indices = np.argsort(eigen_values)[::-1]
    eigen_values = eigen_values[sor_indices]
    eigen_vectors = eigen_vectors[:, sor_indices]

    #selecting the top ones now
    top_eig = eigen_vectors[:, :num_components] #reducing the coloumns to num_components
    
    #calculating mean
    mean = np.mean(data_array , axis = 0)
    
    #calculating the center point
    center = data_array - mean
 
    #multiplying the with top eigen vectors to get pca
    pca_f = np.dot(top_eig.T , center.T).T
    return pca_f

principal_comp = pca_f(data_array , 2)
print(principal_comp)
```
