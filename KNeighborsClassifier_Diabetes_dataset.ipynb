# Diabetes Data Using KneighboursClassifier

```python
#importing the necessary libraries
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
#importing the data set
df = pd.read_csv('c:/csv/diabetes.csv')
#displaying head
df.head()
```

```python
#displaying info of data types
df.info()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#creating bar scatter plot
df.plot(kind='scatter' , x='DiabetesPedigreeFunction' , y='Outcome')
df.plot(kind='box' , x='DiabetesPedigreeFunction' , y='Outcome')
df.plot(kind='hist' , x='DiabetesPedigreeFunction' , y='Outcome')
df.plot(kind='bar' , x='DiabetesPedigreeFunction' , y='Outcome')
```

```python
#dropping the coloumn play in order to predict
X = df.drop('Outcome', axis = 1)
X.head()
```

```python
#creating y data
y = df[['Outcome']]
y.head()
```

```python
#creating train split data
X_train , X_test , y_train , y_test = train_test_split(X , y , random_state = 24, test_size = 0.3)
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
#creating model
model = KNeighborsClassifier(n_neighbors = 5)
model = model.fit(X_train,y_train)
model
```

```python
#predicting the value
y_predict = model.predict(X_test)
y_predict.shape
```

```python
#importing libraries for accuracy
from sklearn.metrics import accuracy_score
print("accuracy_score : ",accuracy_score(y_test , y_predict))
```

```python
#creating confusin matrix
from sklearn.metrics import confusion_matrix
con_mat = confusion_matrix(y_test , y_predict)
con_mat
```

```python
#creating heat map
import seaborn as sns
sns.heatmap(con_mat , annot = True)
```
